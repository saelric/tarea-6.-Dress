```python
import pandas as pd
import numpy as np
```

```python
dress = pd.read_excel("dress.xlsx")
dress.head()
```

```python
dress['Size'].replace('SMALL',"S", inplace=True)
dress['Style']= dress['Style'].str.upper()
cat = pd.Categorical(dress["Style"])
g=1
for f in cat.categories:
    print("Se remplaza "+f+" por "+str(g))
    dress["Style"].replace(f,g, inplace=True)
    g += 1
cat = pd.Categorical(dress["Style"])
cat.describe()

```

```python
dress['Price']= dress['Price'].str.upper()
cat = pd.Categorical(dress["Price"])
g=1
for f in cat.categories:
    print("Se remplaza "+f+" por "+str(g))
    dress["Price"].replace(f,g, inplace=True)
    g += 1
cat = pd.Categorical(dress["Price"])
cat.describe()
```

```python
dress['Size']= dress['Size'].str.upper()
dress['Size'].replace('SMALL',"S", inplace=True)
cat = pd.Categorical(dress["Size"])
g=1
for f in cat.categories:
    print("Se remplaza "+f+" por "+str(g))
    dress["Size"].replace(f,g, inplace=True)
    g += 1
cat = pd.Categorical(dress["Size"])
cat.describe()
```

```python
dress['Season']= dress['Season'].str.upper()
dress['Season'].replace('AUTOMN',"AUTUMN", inplace=True)
cat = pd.Categorical(dress["Season"])
g=1
for f in cat.categories:
    print("Se remplaza "+f+" por "+str(g))
    dress["Season"].replace(f,g, inplace=True)
    g += 1
cat = pd.Categorical(dress["Season"])
cat.describe()
```

```python
dress['SleeveLength']= dress['SleeveLength'].str.upper()
dress['SleeveLength'].replace('CAPSLEEVES',"CAP-SLEEVES", inplace=True)
dress['SleeveLength'].replace('HALFSLEEVE',"HALF", inplace=True)
dress['SleeveLength'].replace('SLEEEVLESS',"SLEEVELESS", inplace=True)
dress['SleeveLength'].replace('SLEEVLESS',"SLEEVELESS", inplace=True)
dress['SleeveLength'].replace('SLEVELESS',"SLEEVELESS", inplace=True)
dress['SleeveLength'].replace('THREEQUATER',"THREEQUARTER", inplace=True)
dress['SleeveLength'].replace('THRESSQATAR',"THREEQUARTER", inplace=True)
dress['SleeveLength'].replace('URNDOWNCOLLOR',"TURNDOWNCOLLOR", inplace=True)
cat = pd.Categorical(dress["SleeveLength"])
g=1
for f in cat.categories:
    print("Se remplaza "+f+" por "+str(g))
    dress["SleeveLength"].replace(f,g, inplace=True)
    g += 1
cat = pd.Categorical(dress["SleeveLength"])
cat.describe()
```

```python
dress['waiseline']= dress['waiseline'].str.upper()
cat = pd.Categorical(dress["waiseline"])
cat.describe()
```

```python
dress['Material']= dress['Material'].str.upper()
cat = pd.Categorical(dress["Material"])
cat.describe()
```
